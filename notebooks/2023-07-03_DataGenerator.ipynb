```python
def get_events(events):
    env = []
    for e in events:
        if "info" not in e:
            t,*m = e.split(":")
            env.append((float(t),m[0].strip()))
    return env
        
    
env = get_events(events)
messages = list(set(y for x,y in env))
markers = ["*","o","x","x","s"]
colors = ["g","k","r","orange","b"]
m = dict([(messages[i], markers[i]) for i in range(len(messages))])
c = dict([(messages[i], colors[i]) for i in range(len(messages))])
```

```python
spikeTimes = d[()]

plt.figure(figsize=(10,5))
plt.scatter(spikeTimes["frame_no"], spikeTimes["channel"], c=spikeTimes["amplitude"], s=1)

for e in env:
    if e[1]=="game begin":
        plt.vlines([e[0]], 0, 102, color="k")
    if e[1]=="ball return":
        plt.vlines([e[0]], 0, 102, color="g")
    if e[1]=="ball missed":
        plt.vlines([e[0]], 0, 102, color="r")
    if "motor" in e[1]:
        plt.vlines([e[0]], 0, 102, color="b")
        
    if e[1]=="ball bounce":
        plt.scatter(e[0], -10, c="b", s=1, label=e[1])
plt.show()
```

```python
all_env = []
for j in joined_data:
    if "event" in j and int(float(j[j.index("events")-2]))%2==0:
        print(j)
        with open(j, "r") as f:
            events = f.readlines()
            env = get_events(events)
            all_env += env
```

```python
successes=[]
for e in all_env:
    if "return" in e[1]:
        if e[0]>1e9:
            successes.append(e[0])
```
